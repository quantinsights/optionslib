```python
import datetime as dt

from optionslib.market.discounting_curve import DiscountingCurve
from optionslib.types.enums import DiscountingInterpolationMethod
```

```python
dates = [
    dt.date(2023, 1, 1),
    dt.date(2024, 1, 1),
    dt.date(2025, 1, 1),
    dt.date(2026, 1, 1),
    dt.date(2027, 1, 1),
    dt.date(2028, 1, 1),
]

dfs = [
    1.00,
    (1.05) ** (-1),
    (1.05) ** (-2),
    (1.05) ** (-3),
    (1.05) ** (-4),
    (1.05) ** (-5),
]
```

```python
discountingCurve = DiscountingCurve(
    dates,
    dfs,
    DiscountingInterpolationMethod.LINEAR_ON_LOG_OF_DISCOUNT_FACTORS,
)
today = dt.date(2023, 1, 1)
df = discountingCurve.discountFactor(today, dt.date(2024, 6, 1))
print(f"DF = {df}")
```

```python
discountingCurve.plotZeroCouponCurve()
```
